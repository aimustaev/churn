## Комбинации разных моделей и для разных препроцессингов

### 1. Все пустышки заполняем нулями

```python
name = 'origin'
models = joblib.load(f'save/prod/models_{name}.pkl')
data = joblib.load(f'save/prod/data_{name}.pkl')
models
```

```python
keras1 = models['KerasRunner{"epochs": 200}']
rf1 = models['RFGridSearch{"n_estimators": [500], "max_features": ["sqrt"], "max_depth": [32], "criterion": ["entropy"]}']

predict1 = keras1.predict(data['X_test'])[:, 1] + rf1.predict_proba(data['X_test'])[:, 1]
print(roc_auc_score(data['y_test'], predict1 / 2))
```
